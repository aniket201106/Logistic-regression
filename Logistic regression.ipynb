```python
#  LOGISTIC REGRESSION:- 

Q1 What is Simple Linear Regression (SLR)? Explain its purpose.
    Simple Linear Regression (SLR) is a statistical method used to study the relationship between one independent variable (X) and one dependent
    variable (Y). It fits a straight line to the data points to predict the value of the dependent variable based on the independent variable.

Purpose:
To understand the relationship between two variables.
To predict future values of the dependent variable.
To identify how changes in the independent variable affect the dependent variable.

For example, predicting a student's marks based on the number of hours studied.

Q2 What are the key assumptions of Simple Linear Regression?
    The key assumptions of Simple Linear Regression are:

1.Linearity – There is a linear relationship between X and Y.
2.Independence – Observations are independent of each other.
3.Homoscedasticity – The variance of errors remains constant for all values of X.
4.Normality of Errors – Residuals (errors) are normally distributed.
5.No Significant Outliers – Extreme values should not strongly influence the regression line.

These assumptions help ensure that the regression model produces reliable results.

Q3 Write the mathematical equation for a simple linear regression model and explain each term?
    The mathematical equation of Simple Linear Regression is:   Y=β0+β1X+ε

Where:
Y = Dependent variable (target/output)
X = Independent variable (predictor/input)
β₀ = Intercept (value of Y when X = 0)
β₁ = Slope or coefficient (change in Y for a one-unit increase in X)
ε = Error term (difference between actual and predicted values)

This equation represents a straight line that best fits the data.

Q4 Provide a real-world example where simple linear regression can be applied.
     A real-world example of Simple Linear Regression is predicting house prices based on the size of the house.

-Independent Variable (X): House size (in square feet)
-Dependent Variable (Y): House price

As the size of the house increases, the price generally increases. Using Simple Linear Regression, we can estimate the price of a house based
on its size.

Another example is predicting sales based on advertising expenditure.

Q5 What is the method of least squares in linear regression?
    The Least Squares Method is a technique used to find the best-fitting regression line.
     It works by minimizing the sum of the squared differences between the actual values and the predicted values.

Mathematically:

Sum of Squared Errors (SSE)=∑(Yi−Yi^)2

Where:

Yi = Actual value
Yi^ = Predicted value

The regression line obtained using the least squares method has the smallest possible total squared error, making it the best fit for the data.

Q6  What is Logistic Regression? How does it differ from Linear Regression?
    Logistic Regression is a supervised machine learning algorithm used for classification problems, where the output is categorical
    (e.g., Yes/No, Pass/Fail, Spam/Not Spam).

Differences between Linear Regression and Logistic Regression:
Linear Regression	                                  Logistic Regression
Used for prediction of continuous values	       Used for classification
Output can be any real number	                   Output is between 0 and 1
Uses a straight-line equation	                   Uses the sigmoid function
Example: House price prediction	                   Example: Email spam detection

Q7 Name and briefly describe three common evaluation metrics for regression models.
    1. Mean Absolute Error (MAE) :- Measures the average absolute difference between actual and predicted values.

MAE=∑∣yi−yi^∣ /n

Lower MAE indicates better performance.


2. Mean Squared Error (MSE) :- Measures the average squared difference between actual and predicted values.

MSE=∑(yi−yi^)2 /n

Penalizes larger errors more heavily.


3. Root Mean Squared Error (RMSE) :- Square root of MSE.

RMSE = √MSE
	
Provides error in the same units as the target variable.

Q8 What is the purpose of the R-squared metric in regression analysis?
    R-squared (R²) measures how well the regression model explains the variability of the dependent variable.

R2=1−SSres/SStot
	
Where:

SSres = Residual Sum of Squares
SStot = Total Sum of Squares

Interpretation:
R² = 1 → Perfect fit
R² = 0 → Model explains no variation
Higher R² → Better model performance

For example, an R² value of 0.85 means the model explains 85% of the variation in the data.
```

```python
# Q9 Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.
#(Include your Python code and output in the code box below.)
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 6, 8, 10])

# Create model
model = LinearRegression()

# Train model
model.fit(X, y)

# Print slope and intercept
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
# Q10 How do you interpret the coefficients in a simple linear regression model?
    In a Simple Linear Regression model:

Y=β0+β1X

Interpretation of Coefficients:

Intercept (β₀)
-Represents the value of Y when X = 0.
-It is the point where the regression line crosses the Y-axis.

Slope (β₁)
-Represents the change in Y for every one-unit increase in X.
-Indicates the strength and direction of the relationship.

Example:
Y=5+3X
Intercept = 5
Slope = 3

Interpretation:

When X = 0, Y = 5.
For every increase of 1 unit in X, Y increases by 3 units.
```
